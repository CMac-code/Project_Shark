```python
# Dependencies
import pandas as pd
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
import requests
import json
import gmaps

# Import API key
from config import g_key

# Configure gmaps
gmaps.configure(api_key=g_key)
```

```python
file="../Resource/Shark Sightings (clean).csv"
sighting_df= pd.read_csv(file)
sighting_df
```

```python
#extract latitude and longitude information from sighting_df
sighting_location=sighting_df[["Latitude","Longitude"]]

# Plot Heatmap
fig = gmaps.figure(layout={'width':'900px','height':'750px'})

# Create heat layer
heat_layer = gmaps.heatmap_layer(sighting_location, 
                                 dissipating=False, max_intensity=1,
                                 point_radius=0.07)


# Add layer
fig.add_layer(heat_layer)

# Display figure
fig
```
